```python
uniuqe_combinations=retail_dataset.groupby(['Member_number','Date']).agg({'itemDescription':list}).reset_index()
uniuqe_combinations
# aggregation on columns or rows of a datarfame to reset index
# .groupby in pandas is often for transaction ID(CUSTOMER) Drouping
```

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
te=TransactionEncoder()
te
# Tansaction Encoder Converts Transaction Data into Frequent itemset
```

```python
te_ary=te.fit_transform(transaction)
te_ary
# Transaction encoder (It is process of converting categorical data into Boolean)
```

```python
dataset=pd.DataFrame(te_ary,columns=te.columns_)
```

```python
dataset
```

```python
# def is a function,encode_unit is a function name
def encode_units(x):
    if x==False:
        return 0
    if x==True:
        return 1
    
```

```python
dataset=dataset.applymap(encode_units)
dataset
```

```python
frequent_itemsets=apriori(dataset,min_support=.10,max_len=2,use_colnames=True)
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=.20)
rules.head(20)
```

```python
frequent_itemsets=apriori(dataset,min_support=.01,max_len=2,use_colnames=True)
rules=association_rules(frequent_itemsets,metric='confidence',min_threshold=0)
rules
```
