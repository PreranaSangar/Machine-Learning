```python
# Dictionary method,NLP(Text Processing) to convert categorical to numerical
d={'UK':0,'USA':1,'N':2}
df[' Nationality']=df[' Nationality'].map(d)
d1={'YES':1,'NO':0}
df['GO']=df['GO'].map(d1)
print(df)
```

```python
df.columns
```

```python
features=['Age', 'Exp', 'Rank', ' Nationality']
```

```python
X=df[features]
```

```python
y=df['GO']
```

```python
y
```

```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree=DecisionTreeClassifier()
dtree.fit(X,y)
```

```python
tree.plot_tree(dtree,feature_names=features)
```
